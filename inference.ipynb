```python
%matplotlib inline

from PIL import Image
from matplotlib import pyplot as plt
from IPython.display import clear_output

from src.models.yolo_model import YoloModel
from src.models.models import VerificationCarsModel
from src.models.cars_model import CarsModel, ArcFace

model_weights = 'weights/tf_efficientnet_b5_drom_v1_4.pth'

car_model = VerificationCarsModel(cars_model_parameters = {},
                                  gpus=[0],
                                  emb_size = 512, 
                                  weights=model_weights)

def draw_images(img_1, img_2, name, fontsize=16):
    
    fig, axes = plt.subplots(1, 2, figsize=(15, 5))
    fig.suptitle(name, fontsize=fontsize)
    axes[0].imshow(img_1)
    axes[1].imshow(img_2);
    
    
img_path_1 = 'files/example3.jpg'
img_path_2 = 'files/example2.jpg'

img_1 = Image.open(img_path_1).convert('RGB')
img_2 = Image.open(img_path_2).convert('RGB')

yolo_model = YoloModel(yolo_name='yolov5s')
res_1 = yolo_model(img_1)
res_2 = yolo_model(img_2)

clear_output()

draw_images(img_1, img_2, "Original Images")

sim = round(car_model(res_1, res_2), 3)
draw_images(res_1, res_2, "Crop Images\nSimilarity={:.3f}".format(sim))
```
